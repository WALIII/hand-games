```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('hand_landmarksmodified1.csv')
thumb = data['thumb']
index = data['index']
middle = data['middle']
ring = data['ring']
pinky = data['pinky']
frame = data['frame']
```

```python
fig = plt.figure()
myaxes = fig.add_axes([1,1,1,1])
myaxes.plot(frame,thumb,'r', lw = 3,label = 'thumb')
myaxes.plot(frame,index,'orange', lw = 3,label = 'index')
myaxes.plot(frame,middle,'y', lw = 3,label = 'middle')
myaxes.plot(frame,ring,'g', lw = 3,label = 'ring')
myaxes.plot(frame,pinky,'b', lw = 3,label = 'pinky')

myaxes.legend()
```
